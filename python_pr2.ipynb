**Завдання 2**

Напишіть програму, яка отримує на вхід слово і виводить True, якщо слово є паліндромом, або False в іншому випадку. Примітка: для порівняння в Python використовується оператор ==.

word[::-1] — створює перевернуту копію слова.

== — порівнює оригінал і перевернуту версію.

Якщо вони однакові — це паліндром, і програма виведе True.

Виконайте приклад нижче:

```python
word = input("Введіть слово: ")

# Перевірка на паліндром
is_palindrome = word == word[::-1]

# Вивід результату
print(is_palindrome)
```

split() розбиває рядок на окремі слова.

capitalize() робить першу літеру великою, а решту — маленькими.

' '.join(...) об'єднує слова назад у один рядок.

Виконайте приклад нижче

```python
# Ввід рядка
full_name = input("Введіть ім'я, по батькові та прізвище: ")

# Розділення на слова та форматування
formatted_name = ' '.join(word.capitalize() for word in full_name.split())

# Вивід результату
print(formatted_name)
```
